#### In the dataset we will consider  neighbourhood, room_type & neighbourhood_group categorical columns for encoding

#### Let us implement Approach#1 i.e., LabelEncoding on neighbourhood column

```python
df['neighbourhood'].value_counts().sort_values(ascending=False)
```

* To implement Label Encoding we've to first convert the column type to category

```python
df['neighbourhood']=df['neighbourhood'].astype('category')
```

```python
df['neighbourhood']=df['neighbourhood'].cat.codes
```

```python
df['neighbourhood'].value_counts()
```

#### Let us implement Approach#2 i.e., Find & Replace on room_type column

```python
df['room_type'].value_counts()
```

```python
df['room_type'].replace({"Private room":1, "Entire home/apt":2,"Shared room":3},inplace=True)
df.head()
```

* Using Find & Replace we've replaced Private Room with 1, Entire home/apt with 2 & Shared roo with 3. Now the value_counts for this column are as below and the datatype of the column is changed from Object to int64:

```python
df['room_type'].value_counts()
```

#### Let us implement Approach#3 i.e., One Hot Encoding on neighbourhood_group

```python
df['neighbourhood_group'].value_counts()
```

```python
df=pd.get_dummies(df,columns=["neighbourhood_group"])
df.head()
```
